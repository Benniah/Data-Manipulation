```python
 
# Using a sample Dataset imported from the Web about Pima Indians , we can do some exploratory analysis and perform
# basic Data analysis using python to give as insights about the Diabetes status of the individuals .
# we will only be using the Pandas and Matplolib libraries




import pandas as pd
import matplotlib.pyplot as plt

#Importing Data Directly From the Web
#df_pidd = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv")


#Importing Data from Desktop 
data = pd.read_csv('diabetesData.csv')
print(data)

#Adding Names To Columns
data.columns = ['NumberOfTimesPregnant' ,'PlasmaGlucose' ,
        'DBP' , 'TricespsThickness' ,'2-HR_Insulin' ,
        'BMI' ,'DiabetesPF' ,'Age' , 'ClassVariable' ]

df_pidd = data

#Find The Number of Size of the File
[nrows, ncols] = df_pidd.shape 


#Print The first 5 rows of the File to check the data types
df_piddHeader = df_pidd.head()
print(df_piddHeader)

#Group the Data according to age and print the nuber of class
print(df_pidd.groupby(["Age","ClassVariable"]).count())


#Filter The Data for  a particular Value of a Variable
Age25 = df_pidd[df_pidd.Age == 25 ].count()
print(Age25)


#Clearing Variables From Varable List

del Age25
del df_piddHeader
del ncols
del nrows

#Calculating The Mean,Min and Maximum of column Age
df_piddMeanAge = df_pidd.Age.mean()
print(df_piddMeanAge)

df_piddMedianAge = df_pidd.Age.median()
print(df_piddMedianAge)

df_piddMaxAge = df_pidd.Age.max()
print(df_piddMaxAge)

df_piddMinAge = df_pidd.Age.min()
print(df_piddMinAge)


#PLOTTING

#Select the Columns Age and Class Values in Subset
#Create a subset of 2 Variables Age and ClassVariable
subset1 = df_pidd[['Age', 'ClassVariable']]


#Using a scatter Plot and plottng values in subset 1
plt.scatter(subset1['Age'], subset1['ClassVariable'])
plt.show()


#Create a subset of Age and Diastolic Blood Pressure
subset2 = df_pidd[['Age','DBP']]

#Sort the Values in Subset2 by Age
subset2.sort_values("Age",inplace = True)


#Plot the Data in subset2
plt.plot(subset2['Age'], subset2['DBP'])
plt.show()
```
